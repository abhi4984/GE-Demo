```python
import pandas as pd
import numpy as np

df = pd.read_csv("Masked_Data_1.csv")
df = df[df['list price'] > 0]
df = df[df['selling Price'] > 0]
pd.set_option("display.max_columns", None)
df
```

```python
df = df.reset_index(drop=True)
month = []
year = []
for i in range(len(df)):
    month.append(str(df['DATE_ORDERED'][i])[3:5])
    year.append(str(df['DATE_ORDERED'][i])[6:10])
for i in range(len(month)):
    month[i] = int(month[i])
    year[i] = int(year[i])
df['month'] = month
df['year'] = year
```

```python
df_nov_dec = pd.read_csv("nov_dec_2020.csv")
df_nov_dec
```

```python
df_jan = df.loc[(df.month==1) & (df.year == 2021)]

cus_jan = df_jan.SHIP_TO_UCMID.unique()
cus_nov_dec = df_nov_dec.SHIP_TO_UCMID.unique()
```

```python
ds1 = pd.DataFrame(columns = ['SHIP_TO_UCMID','BILL_TO_UCMID','self_order','Discount%','DATE_ORDERED','Item Number',
                             'SUB_MODALITY','QUANTITY','list price','selling Price','Total Sale USD','Recency',
                             'Frequency','Monetary','R','F','M','LL','AVG_DIS_CUS','AVG_DIS_ITEM','ORDER_METHOD'
                             ,'AVG_DIS_CUS_ORDER_MET','TOT_SALE_CUS_OFF','TOT_SALE_CUS_ONLINE','month','year'])
ds1
```

```python
for i in range(len(cus_jan)):
    if(cus_jan[i] in cus_nov_dec):
        s = df_jan[df_jan['SHIP_TO_UCMID'] == cus_jan[i]]
        ds1 = ds1.append(s)
        
ds1 = ds1.reset_index(drop = True)
x = []
for i in range(len(ds1)):
    if(ds1['SHIP_TO_UCMID'][i] == ds1['BILL_TO_UCMID'][i]):
        x.append(1)
    else:
        x.append(0)
ds1['self_order'] = x
```

```python
cus_his = ds1.SHIP_TO_UCMID.unique()
cus_his
```

```python
Rec = {}
Freq = {}
Mon = {}
R = {}
F = {}
M = {}
LL = {}
avg_cus_dis = {}
avg_cus_dis_order = {}
tot_sale_cus_off = {}
tot_sale_cus_on = {}
for i in range(len(cus_his)):
    s = df_nov_dec[df_nov_dec['SHIP_TO_UCMID'] == cus_his[i]]
    Rec[cus_his[i]] = s.Recency.unique()[0]
    Freq[cus_his[i]] = s.Frequency.unique()[0]
    Mon[cus_his[i]] = s.Monetary.unique()[0]
    avg_cus_dis[cus_his[i]] = s.AVG_DIS_CUS.unique()[0]
    R[cus_his[i]] = s.R.unique()[0]
    F[cus_his[i]] = s.F.unique()[0]
    M[cus_his[i]] = s.M.unique()[0]
    LL[cus_his[i]] = s.LL.unique()[0]
    avg_cus_dis_order[cus_his[i]] = s.AVG_DIS_CUS_ORDER_MET.unique()[0]
    tot_sale_cus_off[cus_his[i]] = s.TOT_SALE_CUS_OFF.unique()[0]
    tot_sale_cus_on[cus_his[i]] = s.TOT_SALE_CUS_ONLINE.unique()[0]
```

```python
for i in range(len(ds1)):
    ds1['Recency'][i] = Rec[ds1.SHIP_TO_UCMID[i]]
    ds1['Frequency'][i] = Freq[ds1.SHIP_TO_UCMID[i]]
    ds1['Monetary'][i] = Mon[ds1.SHIP_TO_UCMID[i]]
    ds1['R'][i] = R[ds1.SHIP_TO_UCMID[i]]
    ds1['F'][i] = F[ds1.SHIP_TO_UCMID[i]]
    ds1['M'][i] = M[ds1.SHIP_TO_UCMID[i]]
    ds1['LL'][i] = LL[ds1.SHIP_TO_UCMID[i]]
    ds1['AVG_DIS_CUS'][i] = avg_cus_dis[ds1.SHIP_TO_UCMID[i]]
    ds1['AVG_DIS_CUS_ORDER_MET'][i] = avg_cus_dis_order[ds1.SHIP_TO_UCMID[i]]
    ds1['TOT_SALE_CUS_OFF'][i] = tot_sale_cus_off[ds1.SHIP_TO_UCMID[i]]
    ds1['TOT_SALE_CUS_ONLINE'][i] = tot_sale_cus_on[ds1.SHIP_TO_UCMID[i]]
```

```python
item_his = ds1['Item Number'].unique()
len(item_his)
```

```python
item_nov_dec = df_nov_dec['Item Number'].unique()
```

```python
item = {}
for i in range(len(item_his)):
    if item_his[i] in item_nov_dec:
        s = df_nov_dec[df_nov_dec['Item Number'] == item_his[i]]
        item[item_his[i]] = s.AVG_DIS_ITEM.unique()[0]
    else:
        item[item_his[i]] = 0
```

```python
for i in range(len(ds1)):
    ds1['AVG_DIS_ITEM'][i] = item[ds1['Item Number'][i]]
```

```python
ds1['Identifier'] = 'January_2021'
```
